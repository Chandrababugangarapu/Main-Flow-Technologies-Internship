2. Perform simple data cleaning tasks such as  handling missing values and removing  duplicates. 

```python
#1. Handling Missing Values:
#Identifying Missing Values:
# Check for missing values
print(df.isnull().sum())  # Shows the count of missing values per column
```

```python
#dropping Rows with missing values  
df_dropna = df.dropna()  # Drops rows with any missing values (default)
# Alternatively, specify a subset of columns for dropping
df_dropna = df.dropna(subset=["YearsExperience", "Salary"])
df_dropna
```

```python
# Fill missing values with a specific value (e.g., 0)
df["Salary"].fillna(0, inplace=True)  # Fills missing values in column1 with 0

# Fill missing values based on calculations (e.g., mean)
df["Salary"].fillna(df["Salary"].mean(), inplace=True)  # Fills missing prices with mean
```

```python
df_unique = df.drop_duplicates()  # Removes all duplicate rows
df_unique
```

```python
df_unique = df.drop_duplicates(subset=["YearsExperience", "Salary"])  # Considers only column1 & column2
df_unique
```

3. Practice basic data manipulation operationsÂ  like filtering, sorting, and grouping data.

```python
boolean_series = df["YearsExperience"] > 10  # Create a boolean Series based on condition
filtered_df = df[boolean_series]  # Filter using the boolean Series
filtered_df
```

```python
# SORTING DATA
sorted_df = df.sort_values(by="YearsExperience")  # Sort ascending
sorted_df
```

```python
sorted_df = df.sort_values(by="YearsExperience", ascending=False)  # Sort descending
sorted_df
```

```python
sorted_df = df.sort_values(by=["YearsExperience", "Salary"])  # Sort by multiple columns
sorted_df 
```

```python
#Grouping Data:
#Grouping by a single column:
grouped_data = df.groupby("YearsExperience")  # Group DataFrame by column1
grouped_data
```

```python
# Perform operations on groups (e.g., calculate mean)
average_price_per_category = grouped_data["YearsExperience"].mean()
average_price_per_category
```

```python
# Grouping by mutiple columns 
grouped_data = df.groupby(["YearsExperience", "Salary"])  # Group by two columns
grouped_data
```
